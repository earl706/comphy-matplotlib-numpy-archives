```python
import matplotlib.pyplot as plt
import numpy as np
import math

angle = float(input("Enter the starting angle in (degrees): "))
initial_angular_speed = float(input("Enter the initial angular speed (rad/s): "))
angular_acceleration = float(input("Enter the angular acceleration (rad/s^2): "))
radius = float(input("Enter the value for the radius (m): "))
time_of_simulation = float(input("Enter the total time of simulation: "))

angle_to_radians = math.radians(angle)
time = np.linspace(0, time_of_simulation, 1000)

theta_list = [angle_to_radians + initial_angular_speed * t + 0.5 * angular_acceleration * t**2 for t in time]

x_position = [radius*np.cos(theta) for theta in theta_list]
y_position = [radius*np.sin(theta) for theta in theta_list]

plt.figure(figsize=(6, 6))
plt.plot(x_position, y_position, label='Object Path')
plt.scatter(x_position[0], y_position[0], color='red', label='StartPosition') 
plt.scatter(x_position[-1], y_position[-1], color='green', label='Stop Position')
plt.title('Uniform Circular Motion')
plt.xlabel('x position')
plt.ylabel('y position')
plt.grid(True)
plt.legend()
plt.axis('equal') 
plt.show()
```
