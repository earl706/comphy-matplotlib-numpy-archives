```python
first_number_input = input("Enter first number: ")
second_number_input = input("Enter second number: ")
operation = input("Enter operation (+, -, *, /)")

def operate_two_numbers(first_number, second_number, operation):
    try:
        first_number = float(first_number)
        second_number = float(second_number)
        if operation=="+":
            return first_number+second_number
        elif operation=="-":
            return first_number-second_number
        elif operation=="*":
            return first_number*second_number
        elif operation=="/":
            return first_number/second_number
        else:
            return "Invalid Operation"
    except Exception as e:
        return str(e)


result = operate_two_numbers(first_number_input, second_number_input, operation)
print(result)
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def dxdt(x, t):
    dx_dt=x 
    return dx_dt

x0 = 1

t = np.linspace(0, 1, 100)

solution = odeint(dxdt, x0, t)

plt.plot(t, solution)
plt.xlabel("Time t")
plt.ylabel("x(t)")
plt.title(r"Solution of $\frac{dx}{dt} = x$")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


def dNdt(N, t, frequency):
    dN_dt = -frequency*N
    return dN_dt

t = np.linspace(1, 100, 1000)
N = 1000

frequency_1 = 0.1
frequency_2 = 0.3
frequency_3 = 0.5

solution_fq1 = odeint(dNdt, N, t, (frequency_1,))
solution_fq2 = odeint(dNdt, N, t, (frequency_2,))
solution_fq3 = odeint(dNdt, N, t, (frequency_3,))

plt.figure(figsize=(20,15))
plt.plot(t, solution_fq1, label=r"$\lambda = 0.1$", color="red")
plt.plot(t, solution_fq2, label=r"$\lambda = 0.3$", color="green")
plt.plot(t, solution_fq3, label=r"$\lambda = 0.5$", color="blue")
plt.title(r"The solution of $\frac{dN}{dt} = -\lambda N$")
plt.xlabel("Time (t)", fontsize=20)
plt.ylabel("Population (N)", fontsize=20)
plt.grid(True)
plt.legend(fontsize="20")
plt.show()
```
