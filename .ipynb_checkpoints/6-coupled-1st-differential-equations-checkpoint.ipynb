```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def dNdt(N, t, alpha, beta, delta, gamma):
    n1, n2 = N
    dN1_dt = (alpha*n1)-(beta*n1*n2)
    dN2_dt = (delta*n1*n2)-(gamma*n2)
    return [dN1_dt,dN2_dt]

alpha = 0.9
beta = 0.1
delta = 0.075
gamma = 1.5
t = np.linspace(0,10,10)
initial_values = [700, 500]

n1_solution = odeint(dNdt, initial_values, t, (alpha, beta, delta, gamma))
prey = n1_solution[:, 0]
predator = n1_solution[:, 1]

plt.plot(t, prey, label="Prey", color="blue")
plt.plot(t, predator, label="Predator", color="red")
plt.title(r" Coupled First Order Differential Equation")
plt.xlabel("Time (t)", fontsize=20)
plt.ylabel("Population ", fontsize=20)
plt.grid(True)
plt.legend()
plt.show()
```
