```python
import matplotlib.pyplot as plt
import numpy as np
import math


initial_velocity = float(input("Enter the initial velocity (m/s): "))
release_height = float(input("Enter the release height (m): "))
launch_angle = float(input("Enter the launch angle (degrees): "))
gravity = 9.8

degrees_to_radians = math.radians(launch_angle)
time_of_flight = ((initial_velocity*math.sin(degrees_to_radians))+(math.sqrt(((initial_velocity*math.sin(degrees_to_radians))**2)+(2*gravity*release_height))))/gravity
maximum_height = release_height+((initial_velocity**2)*(math.sin(degrees_to_radians)**2))/(2*gravity)
horizontal_range = (initial_velocity*math.cos(degrees_to_radians))*((initial_velocity*math.sin(degrees_to_radians))+(math.sqrt(((initial_velocity*math.sin(degrees_to_radians))**2)+(2*gravity*release_height))))/gravity

time = np.linspace(0,time_of_flight, 100)
print(f"The time of flight is: {round(time_of_flight, 2)} seconds")
print(f"The maximum height is: {round(maximum_height,2)} m")
print(f"The range is: {round(horizontal_range,2)} m")

horizontal_velocity_component = initial_velocity*math.cos(degrees_to_radians)
vertical_velocity_component = initial_velocity*math.sin(degrees_to_radians)

horizontal_distance = horizontal_velocity_component*time
vertical_distance = release_height+vertical_velocity_component * time - 0.5 * gravity * time**2

plt.title("Projectile Trajectory")
plt.ylabel("Vertical Distance (m)")
plt.xlabel("Horizontal Distance (m)")
plt.grid(True)
plt.scatter(horizontal_distance, vertical_distance, label="Projectile Trajectory")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

initial_velocity = 50
gravity = 9.8

time_of_flight = {}
time_of_flight_linspace = {}
vertical_distance = {}
horizontal_distance = {}

angles = [0, 15, 30, 45, 60, 75, 90]

plt.title("Projectile Trajectories for Different Launch Angles")
plt.ylabel("Vertical Distance (m)")
plt.xlabel("Horizontal Distance (m)")
plt.grid(True)

for angle in angles:
    degrees_to_radians = math.radians(angle)
    current_time_of_flight = 2*initial_velocity*math.sin(degrees_to_radians)/gravity
    
    time_of_flight[f"{angle}"] = current_time_of_flight
    time_of_flight_linspace[f"{angle}"] = np.linspace(0,current_time_of_flight,100)
    
    horizontal_velocity_component = initial_velocity*math.cos(degrees_to_radians)
    vertical_velocity_component = initial_velocity*math.sin(degrees_to_radians)
    
    horizontal_distance[f"{angle}"] = horizontal_velocity_component * time_of_flight_linspace[f"{angle}"]
    vertical_distance[f"{angle}"] = vertical_velocity_component * time_of_flight_linspace[f"{angle}"] - 0.5 * gravity * time_of_flight_linspace[f"{angle}"]**2

    plt.plot(horizontal_distance[f"{angle}"], vertical_distance[f"{angle}"], label=f"{angle}")

plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
v0 = 50  # initial velocity in m/s
g = 9.81  # acceleration due to gravity in m/s^2
angles = [0, 15, 30, 45, 60, 75, 90]  # launch angles in degrees

# Time array from 0 to the time it takes to hit the ground
t = np.linspace(0, 10, 500)

# Plotting the trajectories
plt.figure(figsize=(10, 6))

for angle in angles:
    theta = np.radians(angle)  # convert angle to radians
    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2
    
    # Only plot where y >= 0 (the projectile is above the ground)
    plt.plot(x[y >= 0], y[y >= 0], label=f"{angle}°")

# Labels and title
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.title("Projectile Trajectories for Different Launch Angles")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
