```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 1.0
dt = 0.01
t_total = 10.0
n_steps = int(t_total / dt)

# Initial conditions
x0 = 1.0
v0 = 0.0

# Arrays to store time, position, and velocity
time = np.linspace(0, t_total, n_steps)
x = np.zeros(n_steps)
v = np.zeros(n_steps)

# Initial values
x[0] = x0
v[0] = -omega**2 * x[0]

# Verlet integration
x[1] = x[0] + v0 * dt + 0.5 * v[0] * dt**2

for i in range(1, n_steps - 1):
    v[i] = -omega**2 * x[i]
    x[i + 1] = 2 * x[i] - x[i - 1] + v[i] * dt**2

# Plotting the result
plt.plot(time, x, label="Position (Verlet)")
plt.xlabel('Time (s)')
plt.ylabel('Position (x)')
plt.title('Simple Harmonic Oscillator')
plt.legend()
plt.grid()
plt.show()
```
