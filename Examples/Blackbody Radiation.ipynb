```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
h = 6.626e-34  # Planck's constant (J.s)
c = 3.0e8  # speed of light (m/s)
k_B = 1.38e-23  # Boltzmann constant (J/K)

# Function for Planck's Law
def planck(wavelength, T):
    return (2 * h * c**2 / wavelength**5) / (np.exp(h * c / (wavelength * k_B * T)) - 1)

# Wavelength range (in meters)
wavelengths = np.linspace(1e-9, 3e-6, 500)

# Temperatures
temperatures = [3000, 4000, 5000]

# Plot radiation intensity for different temperatures
plt.figure()
for T in temperatures:
    intensity = planck(wavelengths, T)
    plt.plot(wavelengths * 1e9, intensity, label=f'T={T}K')

plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (W/m^2/nm)')
plt.title('Blackbody Radiation Spectrum')
plt.legend()
plt.grid()
plt.show()
```
