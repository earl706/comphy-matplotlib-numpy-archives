```python
import numpy as np
import matplotlib.pyplot as plt

# Fourier series approximation of a square wave
def square_wave(x, n_terms):
    result = np.zeros_like(x)
    for n in range(1, n_terms + 1, 2):  # Odd terms only
        result += (4 / (n * np.pi)) * np.sin(n * x)
    return result

# Parameters
x = np.linspace(0, 2 * np.pi, 500)
n_terms_list = [1, 5, 10, 20]

# Plot Fourier approximations
plt.figure()
for n_terms in n_terms_list:
    y = square_wave(x, n_terms)
    plt.plot(x, y, label=f'{n_terms} terms')

plt.title('Fourier Series Approximation of a Square Wave')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
