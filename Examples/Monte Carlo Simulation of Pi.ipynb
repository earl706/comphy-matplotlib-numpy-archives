```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_points = 10000

# Generate random points in the unit square
x = np.random.rand(n_points)
y = np.random.rand(n_points)

# Check if the points are inside the quarter circle
inside_circle = x**2 + y**2 <= 1.0

# Estimate Pi
pi_estimate = 4 * np.sum(inside_circle) / n_points

# Print the result
print(f'Estimated value of Pi: {pi_estimate}')

# Plot the points
plt.figure(figsize=(6, 6))
plt.scatter(x[inside_circle], y[inside_circle], color='blue', s=1)
plt.scatter(x[~inside_circle], y[~inside_circle], color='red', s=1)
plt.title('Monte Carlo Simulation of Pi')
plt.gca().set_aspect('equal', 'box')
plt.show()
```
