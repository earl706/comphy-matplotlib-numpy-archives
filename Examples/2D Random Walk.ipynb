```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_steps = 1000  # number of steps
n_simulations = 100  # number of simulations

# Function to perform random walk
def random_walk(n_steps):
    steps = np.random.choice([-1, 1], size=(n_steps, 2))
    return np.cumsum(steps, axis=0)

# Simulate random walks
displacements = []
for _ in range(n_simulations):
    trajectory = random_walk(n_steps)
    displacement = np.sqrt(trajectory[:, 0]**2 + trajectory[:, 1]**2)
    displacements.append(displacement)

# Average displacement
avg_displacement = np.mean(displacements, axis=0)

# Plot random walk
plt.plot(avg_displacement, label='Average Displacement')
plt.xlabel('Steps')
plt.ylabel('Displacement from origin')
plt.title('2D Random Walk')
plt.legend()
plt.grid()
plt.show()
```
