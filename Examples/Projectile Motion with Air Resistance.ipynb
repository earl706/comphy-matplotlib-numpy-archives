```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # gravity (m/s^2)
c = 0.1   # drag coefficient
dt = 0.01
t_total = 10.0
n_steps = int(t_total / dt)

# Initial conditions
v0 = 50.0  # initial velocity (m/s)
angle = 45.0  # launch angle (degrees)
vx0 = v0 * np.cos(np.radians(angle))
vy0 = v0 * np.sin(np.radians(angle))

# Arrays to store time, position, and velocity
time = np.linspace(0, t_total, n_steps)
x = np.zeros(n_steps)
y = np.zeros(n_steps)
vx = np.zeros(n_steps)
vy = np.zeros(n_steps)

# Initial values
x[0] = 0.0
y[0] = 0.0
vx[0] = vx0
vy[0] = vy0

# Euler method to solve motion with drag
for i in range(n_steps - 1):
    speed = np.sqrt(vx[i]**2 + vy[i]**2)
    ax = -c * vx[i] * speed
    ay = -g - c * vy[i] * speed
    
    vx[i + 1] = vx[i] + ax * dt
    vy[i + 1] = vy[i] + ay * dt
    
    x[i + 1] = x[i] + vx[i] * dt
    y[i + 1] = y[i] + vy[i] * dt
    
    if y[i + 1] < 0:  # Stop if projectile hits the ground
        break

# Plotting the trajectory
plt.plot(x[:i], y[:i], label='Trajectory with drag')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.title('Projectile Motion with Air Resistance')
plt.legend()
plt.grid()
plt.show()
```
