```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(1,100,100)
acceleration = 3.20
velocity = [t*acceleration for t in time]

plt.title("Time vs Velocity")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.plot(time, velocity)
```

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(1,100,100)
acceleration = 3.20
position = [(1/2)*acceleration*(t**2) for t in time]

plt.title("Time vs Position")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.plot(time, position, label="Time vs Position")
```
